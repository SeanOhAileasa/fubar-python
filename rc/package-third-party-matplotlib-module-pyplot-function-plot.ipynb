- package - matplotlib - unusual in that it breaks a cardinal rule of programming - typically not a good idea for a package to remember internal state - remembering the order in which different functions are called within the package - remembering previous calls [2.2]

- calling the function - plot - the first time - package - matplotlib - creates a new plot [2.2]

```python
plt.plot(nX,nY+nNoise,"r.")
```

- calling the function - plot - the second time - package - matplotlib - checks if a plot was already created - remembers the first plot - sticks on the second plot [2.2]

```python
plt.plot(nX,nY,"b-")
```

- usually not good programming practice for a package to remember a previous call to a function [2.2]

- package - matplotlib - does this for a good reason - good use case - for interactive plotting everyone agrees its ok [2.2]

- function - show - besides showing the figure - also cleans the slate - plotted axes are wiped - in a way implements the whole plot - with all the different settings via functions like - title - xlabel - ylabel - legend - when called - everything is wiped clean [2.2]

````python
# package numpy - subpackage core - C function arange alias arrayEvenlySpaced - call - constant pi - access
nX=arrayEvenlySpaced(-2.0*pi,2.0*pi,0.1)
````

````python
# package numpy - subpackage core - C function sine alias sine - call
nSine=sine(nX)
````

````python
# package matplotlib - module pyplot - function plot alias p - call
p(nX,nSine,"g.")
````

````python
# package numpy - subpackage core - C function cos alias cosine - call
nCos=cosine(nX)
````

````python
# package matplotlib - module pyplot - function plot alias p - call
p(nX,nCos,"b.")
````

- plotting for different values of - ``x`` - sine - cosine - between - $2*\pi$ - and - $-2*\pi$ - on same set of axes - called sinusoids - periodic structure - appear exactly the same - one is shifted slightly to the right [2.2] 

````python
# package matplotlib - module pyplot - function plot alias p - call
p(nX,nSine,"g.")
p(nX,nCos,"b.")
````

- plotting for powers of - ``nX`` - green line - $d^2$ - blue line - $d^3$ - red line - $d^4$ - black line - $2^x$ - appears to be growing the slowest [2.2]

````python
# package numpy - subpackage core - C function arange alias arrayEvenlySpaced - call
nX=arrayEvenlySpaced(1.0,10.0,0.1)
````

````python
# package matplotlib - module pyplot - function plot alias p - call
p(nX,nX**2,"g-",label="x^2")
p(nX,nX**3,"b-",label="x^3")
p(nX,nX**4,"r-",label="x^4")
p(nX,2**nX,"k-",label="2^x")
# package matplotlib - module - pyplot - function legend alias ldg - call
lgd()
````

````python
# package numpy - subpackage core - C function arange alias arrayEvenlySpaced - call
nX=arrayEvenlySpaced(1.0,100.0,0.1)
````

- changing - axis - ``x`` - name - ``nX`` - go to - ``100`` - black line - $2^x$ - dominates - exponential growth - cannot differentiate the other lines given they are so close to the axis - ``x`` [2.2]

````python
# package matplotlib - module pyplot - function plot alias p - call
p(nX,nX**2,"g-",label="x^2")
p(nX,nX**3,"b-",label="x^3")
p(nX,nX**4,"r-",label="x^4")
p(nX,2**nX,"k-",label="2^x")
# package matplotlib - module - pyplot - function legend alias ldg - call
lgd()
````
