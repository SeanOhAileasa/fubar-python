```python
from numpy.random import default_rng as newGenBitGenPCG64
from numpy.random import RandomState as CLegacyGenerator
from numpy.random import MT19937 as CSlowMersenneTwister
```

```python
# package matplotlib - module pyplot - function scatter alias s - copy - attribute __doc__ - access
print(s.__doc__)
```

- plotting 2-dimensional data - function - plot - same result as - function - scatter [2.2]

- function - scatter - has extra features - plotting three or four dimensional data - i) axis - ``x`` - name - ``nX`` - ii) axis - ``y`` - name - ``nY`` - iii) dot - ``colour`` - name - ``nColor`` - iv) dot - ``shape`` - name - ``nShape`` [2.2]

```python
# package numpy - subpackage random - object Generator - method default_rng alias newGenBitGenPCG64 - instantiate
nInsObjGBT=newGenBitGenPCG64()
```

```python
# package numpy - subpackage random - object Generator - method uniform - arguments
nLowX,nHighX,nSizeX=0.0,10.0,100
# package numpy - subpackage random - object Generator - method uniform - call
nX=nInsObjGBT.uniform(nLowX,nHighX,nSizeX)
```

```python
# package numpy - subpackage random - object Generator - method uniform - arguments
nLowY,nHighY,nSizeY=0.0,100.0,100
# package numpy - subpackage random - object Generator - method uniform - call
nY=nInsObjGBT.uniform(nLowY,nHighY,nSizeY)
```

```python
# package numpy - subpackage random - object Generator - method normal - arguments
nMu,nSigma,nSize=100.0,0.4,100
# package numpy - subpackage random - object Generator - method normal - call
nShape=nInsObjGBT.normal(nMu,nSigma,nSize)
```

```python
# package numpy - subpackage random - object MT19937 alias CSlowMersenneTwister - instantiate
nMT19937=CSlowMersenneTwister()
# package numpy - subpackage random - object RandomState alias CLegacyGenerator - instantiate
nInsObjLegacyGen=CLegacyGenerator(nMT19937)
# package numpy - subpackage random - object RandomState - method randint - arguments
nLow,nHigh,nSize=0,20,100
# package numpy - subpackage random - object RandomState - method randint - call
nColor=nInsObjLegacyGen.randint(nLow,nHigh,nSize)
```

```python
# package matplotlib - module pyplot - function scatter alias s - call
s(nX,nY,s=nShape,c=nColor)
```
